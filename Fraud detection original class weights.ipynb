```python
df.drop_duplicates(inplace=True,ignore_index=True)
```

# Data Preprocessing

```python
x = df.drop('Class', axis= 1)
y = df['Class']
```

```python
y.value_counts()
```

```python
y.value_counts().round()/len(df)*100
```

* ### Split Data into Train & Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42, stratify= y)
```

* Handle numerical columns

```python
scalin_col=x_train.columns
scalin_col
```

```python
from sklearn.preprocessing import RobustScaler

rc=RobustScaler()

x_train[scalin_col]=rc.fit_transform(x_train)
x_test[scalin_col]=rc.fit_transform(x_test)
```

```python
x_train
```

# Machine Learning

* Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report


lr=LogisticRegression(random_state= 42, n_jobs= -1,class_weight= 'balanced')
lr.fit(x_train,y_train)

print('\nTrain LR :\n', classification_report(y_train, lr.predict(x_train), target_names=['Not Fraud', 'Fraud']))
print('\nTest LR :\n', classification_report(y_test, lr.predict(x_test), target_names=['Not Fraud', 'Fraud']))
```

* Descion Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state= 42,class_weight= 'balanced')

dt.fit(x_train, y_train)

print('\nTrain dt :\n', classification_report(y_train, dt.predict(x_train),target_names=['Not Fraud', 'Fraud']))
print('\nTest dt :\n', classification_report(y_test, dt.predict(x_test),target_names=['Not Fraud', 'Fraud']))
```

* Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state= 42, n_jobs= -1,class_weight= 'balanced')

rf.fit(x_train, y_train)

print('\nTrain rf :\n', classification_report(y_train, rf.predict(x_train),target_names=['Not Fraud', 'Fraud']))
print('\nTest rf :\n', classification_report(y_test, rf.predict(x_test),target_names=['Not Fraud', 'Fraud']))
```

* xgbboost

```python
from xgboost import XGBClassifier


xgb = XGBClassifier(random_state = 42,n_jobs=-1,class_weight= 'balanced')

xgb.fit(x_train, y_train)

print('\nTrain xgb :\n', classification_report(y_train, xgb.predict(x_train)))
print('\nTest xgb :\n', classification_report(y_test, xgb.predict(x_test)))
```
